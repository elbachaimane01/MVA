## 1.2. Model presentation 

### 1.2.1. Graph Isomorphism Network (GIN)

Graph Isomorphism Network (GIN) is a type of Graph Neural Network (GNN) that has been proven to be the most expressive among the class of GNNs. This architecture was introduced in the paper "How Powerful are Graph Neural Networks?" by Xu et al. (2018). 

GIN networks are designed to capture the structural information of graphs and learn their representations. They use a simple yet powerful message-passing scheme that aggregates information from neighboring nodes and edges to update node features. The key idea behind GIN is to use a permutation-invariant aggregation function, which ensures that the order of nodes does not affect the final representation of the graph. This makes GIN networks highly effective in capturing graph structures and achieving state-of-the-art performance on many graph classification benchmarks.

One of the main perks of GIN networks is their high representational power, which allows them to almost perfectly fit training data. This is due to their ability to capture complex structural information from graphs, such as subgraph patterns and cycles, which other GNN variants may struggle with. Additionally, GIN networks are computationally efficient and can handle large-scale graphs with millions of nodes and edges.

Studying GIN networks is important because it provides insights into the limitations and capabilities of GNNs for graph representation learning. By understanding how different architectures perform on various graph classification tasks, we can develop more effective models for real-world applications such as drug discovery, social network analysis, and recommendation systems.

In conclusion, Graph Isomorphism Networks are a powerful type of Graph Neural Network that can effectively capture complex structural information from graphs. Their high representational power and computational efficiency make them an important area of study for researchers interested in graph representation learning.

### 1.2.2 GraphSAGE

GraphSAGE is a framework for inductive representation learning on large graphs, introduced in the paper "Inductive Representation Learning on Large Graphs" by William L. Hamilton, Rex Ying, and Jure Leskovec. 

GraphSAGE is designed to generate node embeddings for previously unseen data by leveraging node feature information. This approach is more efficient and generalizable than previous methods that require all nodes to be present during training. 

The key idea behind GraphSAGE is to learn a function that generates embeddings by sampling and aggregating features from a node's local neighborhood. Instead of training individual embeddings for each node, GraphSAGE learns a function that generalizes to unseen nodes. By incorporating node features in the learning algorithm, GraphSAGE simultaneously learns the topological structure of each node's neighborhood as well as the distribution of node features in the neighborhood.

One of the main advantages of GraphSAGE is its ability to handle feature-rich graphs (e.g., citation data with text attributes, biological data with functional/molecular markers), while also making use of structural features that are present in all graphs (e.g., node degrees). This makes it a powerful tool for a wide range of applications, including social network analysis, recommendation systems, and bioinformatics.

To generate embeddings using GraphSAGE, the framework first samples a fixed number of neighbors for each node in the graph. It then aggregates feature information from these neighbors using a neural network architecture that can be customized based on the specific application. Finally, it generates an embedding for each node based on this aggregated information.

One important aspect of GraphSAGE is its ability to handle large-scale graphs efficiently. By sampling only a fixed number of neighbors for each node and aggregating their features using a neural network architecture that can be trained efficiently using stochastic gradient descent (SGD), GraphSAGE can generate embeddings for large graphs in a reasonable amount of time.

In summary, GraphSAGE is a powerful framework for inductive representation learning on large graphs that leverages node feature information to generate embeddings for previously unseen data. Its ability to handle feature-rich graphs and its efficiency make it an important tool for a wide range of applications.

## 1.3. Data presentation 

### 1.3.1 Reddit binary for graph classification

Introduced by Yanardag et al. in Deep Graph Kernels
REDDIT-BINARY consists of graphs corresponding to online discussions on Reddit. In each graph, nodes represent users, and there is an edge between them if at least one of them respond to the others comment. There are four popular subreddits, namely, IAmA, AskReddit, TrollXChromosomes, and atheism. IAmA and AskReddit are two question/answer based subreddits, and TrollXChromosomes and atheism are two discussion-based subreddits. A graph is labeled according to whether it belongs to a question/answer-based community or a discussion-based community.



Introduced by Dwivedi et al. in Benchmarking Graph Neural Networks
PATTERN is a node classification tasks generated with Stochastic Block Models, which is widely used to model communities in social networks by modulating the intra- and extra-communities connections, thereby controlling the difficulty of the task. PATTERN tests the fundamental graph task of recognizing specific predetermined subgraphs.

### 1.3.2. TSP dataset for Edge Classification

Introduced by Pouya et al. in A new benchmark set for Traveling salesman problem and Hamiltonian cycle problem
This is a benchmark set for Traveling salesman problem (TSP) with characteristics that are different from the existing benchmark sets. In particular, it focuses on small instances which prove to be challenging for one or more state-of-the-art TSP algorithms. These instances are based on difficult instances of Hamiltonian cycle problem (HCP). This includes instances from literature, specially modified randomly generated instances, and instances arising from the conversion of other difficult problems to HCP.

## 1.4. EVALUATION FRAMEWORK

### 1.4.1. A Fair Comparison of Graph Neural Networks for Graph Classification

1. External k-out-fold cross-validation: This step involves splitting the dataset into k equally sized folds, and then training and evaluating the model k times. In each iteration, one of the folds is used as a validation set, while the remaining k-1 folds are used for training. This process is repeated until each fold has been used as a validation set exactly once. The results of each iteration are then averaged to obtain an estimate of the model's generalization performance.

2. Hold-out technique: In this step, a separate hold-out set is created from the original dataset, which is not used during cross-validation. The hold-out set is used to select the best hyperparameters for the model based on its performance on this set.

3. Inner k-inn-fold cross-validation: This step involves applying a complete nested cross-validation to select hyperparameters for the model. This means that in addition to using an external k-out-fold cross-validation (as described in step 1), an inner k-inn-fold cross-validation is applied within each fold of the external cross-validation loop. This process can be computationally expensive since it involves training and evaluating multiple models with different hyperparameters.

Overall, this evaluation framework provides a rigorous and reproducible way to evaluate machine learning models for graph classification tasks.

# 3 . Task 1: Graph classification 

In the field of graph analysis, graph classification is an essential task that involves predicting a label or category for an entire graph rather than just individual nodes or edges. In this section of the code, we will focus on the 'A Fair Comparison of Graph Neural Networks for Graph Classification' paper and provide code to compare two popular graph neural network models, GIN and GraphSAGE, on a social dataset called Reddit-B. GIN and GraphSAGE are both powerful models for graph classification tasks, but they differ in the way they process graph data. While GIN aggregates node features by summing them up and then applying a neural network layer, GraphSAGE uses a message passing mechanism to propagate node features throughout the graph. The Reddit-B dataset is a large-scale social network dataset that consists of approximately 230,000 graphs, where each graph represents a discussion thread on Reddit. In this section, we will use this dataset to compare the performance of GIN and GraphSAGE on graph classification tasks, and explore the strengths and weaknesses of each model.



In the paper, the authors evaluate their Graph Neural Network models using two different settings: with and without node features. 

In the setting with node features, each node in the graph is associated with a feature vector that describes its attributes. These features are used as input to the Graph Neural Network model, which then learns to predict the correct label for each graph.

In contrast, in the setting without node features, no feature vectors are provided as input to the model. Instead, each node is represented only by its connectivity information (i.e., which other nodes it is connected to in the graph). The Graph Neural Network model must then learn to extract useful information from this connectivity information alone in order to make accurate predictions.

By evaluating their models in both settings, the authors are able to assess how well their models can generalize to new graphs when given different types of input information. This is an important consideration since not all graphs may have meaningful or easily extractable node features.
